```python
import pandas as pd

csup = pd.read_stata("D:\\datasets\\coorte13_ces_enem.dta")

df_micro = pd.read_pickle("C:\\Users\\erugg\\OneDrive\\Documentos\\RA\\highereducation_networks\\df_micro.pkl")
```

```python
# Ensure co_curso_list is a list type (in case it's stored as a string in df_micro)
df_micro['co_curso_list'] = df_micro['co_curso_list'].apply(lambda x: tuple(sorted(x)))  # Convert to tuple for uniqueness

# Get unique lists of co_curso
unique_course_lists = df_micro['co_curso_list'].unique()

# Dictionary to store counts
course_list_counts = {}

# Iterate over each unique list of co_curso
for course_list in unique_course_lists:
    # Filter csup by the entire list
    count = csup[csup['co_curso'].isin(course_list)].shape[0]
    course_list_counts[course_list] = count

# Convert to DataFrame for better visualization
course_list_counts_df = pd.DataFrame(list(course_list_counts.items()), columns=['co_curso_list', 'count'])

```
